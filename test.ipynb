```python
from ultralytics import YOLO
import os

model = YOLO('D:\\LPR\\model_- 12 january 2024 14_47.pt')

detection_results_list = []

# Iterate through each image in the folder
for img_path in img_paths:
    results = model(img_path)

    # Extract information and format as desired
    image_detections = []
    for r in results:
        if len(r.boxes.xyxy) > 0:
            detection_result = {
                "name": r.names[0],
                "class": r.boxes.cls[0].item(),
                "confidence": r.boxes.conf[0].item(),
                "box": {
                    "x1": r.boxes.xyxy[0][0].item(),
                    "y1": r.boxes.xyxy[0][1].item(),
                    "x2": r.boxes.xyxy[0][2].item(),
                    "y2": r.boxes.xyxy[0][3].item(),
                }
            }
            image_detections.append(detection_result)

    detection_results_list.append(image_detections)

# Now detection_results_list contains a list of formatted detection results for each image
print(detection_results_list)
```

```python
import cv2
import pytesseract

# Set the path to the Tesseract executable (modify it based on your installation)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# List of image paths
img_paths = img_paths

# List of detection results for each image
detection_results_list = detection_results_list

# Iterate through each image and its corresponding detection results
for img_path, detection_results in zip(img_paths, detection_results_list):
    # Load the image
    img = cv2.imread(img_path)

    # Check if the image is successfully loaded
    if img is None:
        print(f"Error: Unable to read image {img_path}")
    else:
        # Assuming you have a list of detection results
        # Extract bounding box coordinates
        result = detection_results[0]  # Assuming only one result for simplicity
        x1, y1, x2, y2 = result["box"]["x1"], result["box"]["y1"], result["box"]["x2"], result["box"]["y2"]

        # Extract the license plate region from the image
        license_plate = img[int(y1):int(y2), int(x1):int(x2)].copy()

        # Convert the license plate region to grayscale
        license_plate_gray = cv2.cvtColor(license_plate, cv2.COLOR_BGR2GRAY)

        # Threshold the license plate image
        _, license_plate_thresh = cv2.threshold(license_plate_gray, 64, 255, cv2.THRESH_BINARY_INV)

        # Perform OCR using PyTesseract on the thresholded license plate image
        text = pytesseract.image_to_string(license_plate_thresh, config='--psm 8')

        # Process the detected text
        if text.strip():
            print(f"Detected Text in {img_path}:", text.strip())
        else:
            print(f"No text detected in the license plate region of {img_path}")
```
